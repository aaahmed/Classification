```python
#Choose arrays for regression analysis
X = dataset.iloc[:, [0, 1,2,7,8,9,10,11,12]].values
y = dataset.iloc[:, -1].values
```

```python
pd.DataFrame(X)
```

```python
#since data exist is different formats 
#we need to find a common standard using standard scaler
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
# Splitting dataset into training and test arrays
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 41)

```

```python
# Feature Scaling to avoid dicrepencies between inputted X values
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Confirm results
X_train, X_test
```

```python
# Fitting training set to Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 42)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
#print results
cm
```

```python
#Calculate Accuracy
from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
print(cr)
```
